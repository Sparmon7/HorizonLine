```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import csv
from tensorflow.keras import layers, models
```

```python
metadata={}
with open('data/metadata.csv', 'r', newline='') as f:
    reader = csv.reader(f)
    for row in reader:
        values = np.array(row[1:5], dtype = np.float32) 
        metadata[row[0]] = values
```

```python
from datetime import datetime

X_train=[]
with open('data/split/train.txt','r') as file:
    for line in file:
        X_train.append(line.split()[0])
```

```python
train = tf.data.Dataset.from_tensor_slices(X_train)

def load_and_preprocess_image(path):
    image = tf.io.read_file("data/images/"+path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.cast(image, tf.float32)
    image = image / 255.0
    return image

def get_label(path):
    key = path.numpy().decode()
    return tf.constant(metadata[key], dtype=tf.float32)

def process_path(path):
    image = load_and_preprocess_image(path)
    label = tf.py_function(get_label, [path], tf.float32)
    label.set_shape([4])
    return image, label

BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

dataset = train.map(process_path, num_parallel_calls=AUTOTUNE)
dataset = dataset.shuffle(1000).batch(BATCH_SIZE).prefetch(AUTOTUNE)
```

```python
def plot_with_horizon(path):
    data = metadata[path].copy()
    im=load_and_preprocess_image(path)
    sz =im.shape[:2]
    x_shift = (sz[1] + 1) / 2
    y_shift = (sz[0] + 1) / 2
    plt.xlim([1 - (sz[1] + 1) / 2, sz[1] - (sz[1] + 1) / 2])
    plt.ylim([sz[0] - (sz[0] + 1) / 2, 1 - (sz[0] + 1) / 2])
    plt.imshow(im, extent=[1 - x_shift, sz[1] - x_shift, sz[0] - y_shift, 1 - y_shift])
    plt.plot([data[0], data[2]], [data[1], data[3]], 'g-', linewidth=3)
    plt.show()
for i in np.random.randint(0, len(X_train), size=5):
    plot_with_horizon(X_train[i])
```

```python
model = models.Sequential([ 
    layers.InputLayer(shape=(224, 224, 3)), #MUST CHANGE SIZE
    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(4)
])
model.compile(optimizer='adam', loss='mse')

#model.fit(dataset, epochs=10)
```
